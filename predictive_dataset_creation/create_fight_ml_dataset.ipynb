```python
# 1. Imports and file paths
import pandas as pd
import numpy as np

# File paths
FIGHTS_PATH = "../build/fights_classified.csv"
PROFILES_PATH = "../build/fighter_profiles.csv"
OUTPUT_PATH = "../build/fight_ml_dataset.csv"

```

```python
# 2. Load data
fights = pd.read_csv(FIGHTS_PATH)
profiles = pd.read_csv(PROFILES_PATH)

print(f"Loaded {len(fights)} fights and {len(profiles)} fighter profiles.")
display(fights.head())
display(profiles.head())
```

```python
print(fights.columns)
fights.head()
```

```python
# Identify fighter ID columns in fights_unified.csv
print(fights.columns)

# For this example, let's assume columns are named 'fighter_A_id' and 'fighter_B_id'.
# If not, update the column names below accordingly.
A_ID_COL = 'fighter_a_url'
B_ID_COL = 'fighter_b_url'
profile_cols = [col for col in profiles.columns if col != 'fighter_id']

# Merge for Fighter A
rename_dict_A = {col: f"{col}_A" for col in profile_cols}
rename_dict_A['fighter_id'] = A_ID_COL
profiles_A = profiles.rename(columns=rename_dict_A)
fights = fights.merge(profiles_A, on=A_ID_COL, how='left')

# Merge for Fighter B
rename_dict_B = {col: f"{col}_B" for col in profile_cols}
rename_dict_B['fighter_id'] = B_ID_COL
profiles_B = profiles.rename(columns=rename_dict_B)
fights = fights.merge(profiles_B, on=B_ID_COL, how='left')

print(f"After merging: {fights.shape}")
display(fights.head())
print("Null vakues") 
print(fights.isna().sum().sort_values(ascending=False).head(20))
```

```python
# List of numeric features to create differences for
numeric_features = [
    'current_elo_rating', 'age', 'reach_inches', 'height_inches', 'weight_lbs',
    'win_rate', 'total_fights', 'total_wins', 'total_losses', 'total_draws'
]

for feat in numeric_features:
    col_A = f"{feat}_A"
    col_B = f"{feat}_B"
    diff_col = f"{feat}_diff"
    if col_A in fights.columns and col_B in fights.columns:
        fights[diff_col] = fights[col_A] - fights[col_B]

print(f"After difference features: {fights.shape}")
fights.sort_values('DATE', ascending=False).head()
```

```python
## for future find a way to kep drawand no contest
print(fights['winner_label'].value_counts())
# Create target column
# Step 1: Remove rows where winner_label is 'draw' or 'nc'
fights = fights[~fights['winner_label'].isin(['draw', 'nc'])]

# Step 2: Check the remaining unique values (should be 'A' and 'B')
print(fights['winner_label'].value_counts())

# Step 3: Create binary target column: A=1, B=0
fights['target'] = (fights['winner_label'] == 'A').astype(int)

# Step 4: Confirm target column
print(fights['target'].value_counts())
print(fights[['winner_label', 'target']].head())
```

```python
missing = fights.isna().sum()
print("Before imputation")
print(missing[missing > 0])

# Impute missing numeric values with median (no extra columns)
for col in fights.columns:
    if fights[col].dtype in [np.float64, np.int64]:
        if fights[col].isna().any():
            fights[col] = fights[col].fillna(fights[col].median())

# Check for any remaining missing values
missing = fights.isna().sum()
print("After imputation")
print(missing[missing > 0])
```

```python
# Final check: all features numeric except for fight metadata columns
non_numeric = fights.select_dtypes(include=['object']).columns.tolist()
print("Non-numeric columns:", non_numeric)

# Save ML dataset (drop metadata columns if desired)
ml_cols = [col for col in fights.columns if col not in ['winner', 'fight_id', 'event', 'date']]  # keep only features and target
fights[ml_cols].to_csv(OUTPUT_PATH, index=False)
print(f"Saved ML-ready dataset to {OUTPUT_PATH}")

display(fights[ml_cols].head())
```
